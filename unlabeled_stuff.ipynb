```python
import pandas as pd
df_siam = pd.read_csv('data/up/scores/SiamX.csv', index_col=0)
```

```python
labels = [int(i in x) for i in df_siam.index]
```

```python
df_siam.insert(5, 'label', labels)
```

```python
df_siam.to_csv('data/up/SiamX_labeled.csv')
```

```python
from imblearn.under_sampling import CondensedNearestNeighbour
import numpy as np
cnn = CondensedNearestNeighbour(random_state=42)
X = np.reshape(df_siam.score.array, (-1, 1))
y = df_siam.label
X_re, y_re = cnn.fit_resample(X, y)
```

```python
import matplotlib.pyplot as plt
pos = X[y == 1]
neg = X[y == 0]
plt.scatter(pos, [1]*len(pos), marker='o')
plt.scatter(neg, [0]*len(neg), marker='o')
plt.show()
```

```python
import matplotlib.pyplot as plt
pos = X_re[y_re == 1]
neg = X_re[y_re == 0]
plt.scatter(pos, [1]*len(pos), marker='o')
plt.scatter(neg, [0]*len(neg), marker='o')
plt.show()
```

```python
index = df_siam.index 
samples = [index[i] for i in cnn.sample_indices_]
```

```python
df_samples = df_siam.loc[samples]
```

```python
df_samples.to_csv('data/up/SiamX_sampled.csv')
```

```python
import pandas as pd
import pickle
conf_up = {}
all = set()
for name in ['LCS', 'COS', 'LEV', 'LSH', 'D2VK', 'WMDK', 'SiamX2']:
    df = pd.read_csv('data/up/scores/' + name + '.csv', index_col=0, na_filter=False)
    pos = len(df.index)
    with open('data/up/accepted/'+name+'.pkl', 'rb') as f:
        x = pickle.load(f)
        acc = len(x)
        all.update(x)
    conf_up[name] = {'tp': acc, 'fp': pos - acc}
with open('data/up/accepted/all.pkl', 'wb') as f:
    pickle.dump(all, f)
    print(len(all))
conf_up
```

```python
conf = {}
conf['LCS'] = {'tp': 348, 'tn': 359, 'fp': 9, 'fn': 20}
conf['COS'] = {'tp': 346, 'tn': 358, 'fp': 10, 'fn': 22}
conf['LEV'] = {'tp': 357, 'tn': 338, 'fp': 30, 'fn': 11}
conf['LSH'] = {'tp': 334, 'tn': 358, 'fp': 10, 'fn': 34}
conf['D2VK'] = {'tp': 367, 'tn': 346, 'fp': 22, 'fn': 1}
conf['WMDK'] = {'tp': 348, 'tn': 361, 'fp': 7, 'fn': 20}
conf['SiamX2'] = {'tp': 358, 'tn': 336, 'fp': 32, 'fn': 10}

for _, c in conf.items():
    tp = c['tp']
    fp = c['fp']
    fn = c['fn']
    prec = tp / (tp + fp)
    recall = tp / (tp + fn)
    f1 = 2*prec*recall/(prec+recall)
    print(f'{name} {prec:.2f} {recall:.2f} {f1:.2f}')
```

```python
p = 234 + 368
for name in conf.keys():
    tp = conf[name]['tp'] + conf_up[name]['tp']
    fp = conf[name]['fp'] + conf_up[name]['fp']
    prec = tp / (tp + fp)
    recall = tp / p
    f1 = 2*prec*recall/(prec+recall)
    print(f'{name} {tp} {fp} {prec:.2f} {recall:.2f} {f1:.2f}')
```
